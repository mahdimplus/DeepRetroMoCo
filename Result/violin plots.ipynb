```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

```

# direction 

```python
direction='/media/mahdi/HD710 PRO/R&d/result/tsnr_centerfix_result.csv'
df = pd.read_csv(direction) 
df
```

# seperate spine ans csf part

```python
csf_part=df.iloc[0:5,:]
csf_part
spine_part=df.iloc[5::,:]
spine_part
```

# change order of data for violin

```python
csf_part=csf_part.melt(id_vars=["methods"], 
        var_name="subject_ID", 
        value_name="Value")

spine_part=spine_part.melt(id_vars=["methods"], 
        var_name="subject_ID", 
        value_name="Value")
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20, 10))
ax=sns.violinplot(x="methods" ,y="Value", data=csf_part)
fig = ax.get_figure()
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)


plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('tSNR report of CSF part',fontsize=20)


fig.savefig('csf.jpeg')
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20, 10))
ax1=sns.violinplot(x="methods", y="Value", data=spine_part)
fig = ax1.get_figure()
ax1.xaxis.label.set_size(20)
ax1.yaxis.label.set_size(20)


plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('tSNR report of Spine part',fontsize=20)
fig.savefig('spine.jpeg')
```
